# <a name="sec1"></a>  1- Backstripping equation


<img src="figs/ex6-fig1.jpeg" width="100%">

As a reminder, the total subsidence of the basement underlying a sedimentary basin is made of 3 parts at any time $t$: the sediment deposited $y(t)$, the water depth $W_d(t)$, and the change in global sea level $\Delta_{SL}(t)$. We can perform an isostatic balance between a lithospheric column through the sedimentary basin, and a column in which the sediment load has been removed and has been replaced with water (Fig. 7) to obtain the tectonic or backstripped subsidence $Y(t)$. The isostatic balance down to a depth of compensation in the ductile asthenosphere is

$$ \rho_wgW_d(t) + \overline{\rho_b}gy(t) + \rho_cgy_c +  \rho_mga_1 =  \rho_wgY(t) + \rho_cgy_c  + \rho_mga_2 \;\;\;(1)$$ 

where $y_c$ is the crustal thickness, and $a_1, a_2$ are the mantle lithosphere thickness before and after removal of the sediment load, and $\rho_w,\overline{\rho_b},\rho_c, \rho_m$ are the densities of water, bulk sediment, crust and mantle respectively.

Global sea-level change $\Delta_{SL}(t)$ changes the datum of the top of the lithospheric column. Equating the thicknesses in the 2 columns, and removing $y_c$ since crustal thickness does not change, and $g$ since it occurs in each term, we have

$$ W_d(t) - \Delta_{SL}(t) + y(t) + a_1 =  Y(t) + a_2 \;\;\;(2)$$

# <a name="sec2"></a> 2- Sediment bulk density

Assuming the basin to be composed of a **uniform lithology**, the average basin porosity is the integration of the porosity over the depth range of the entire basin-fill thickness $y$, which when evaluated gives 

$$  \overline{\phi} = \frac{\phi_0 y_0}{y}(1-exp(-y/y_0))$$

where $y_0$ is the depth scale for the exponential reduction of porosity and $\phi_0$ is its surface value.

*** 

### Demonstration

Athy's law:

$$ \phi = \phi_0 exp(-y/y_0) $$

Integration of the above equation is defined by:

$  \int \phi dy = \int \phi_0 exp(-y/y_0) dy $

We perform an u-substitution of the variable y such that $u = -y/y_0$ which gives $du = -dy/y_0$ and $dy = -y_0 du$

$ \int  \phi_0 exp(-y/y_0) dy = \int -\phi_0  exp(u) y_0 du $

$ -\phi_0  y_0 \int exp(u) du = -\phi_0  y_0 exp(u) + C $

$ \int \phi dy = -\phi_0  y_0 exp(-y/y_0) + C $

Average porosity is then computed by dividing the integration of the porosity over the basin thickness:

$  \overline{\phi} = \frac{\left[ -\phi_0  y_0 exp(-y/y_0) + C \right]}{y} $

Value of C is given from the limit $y=0$, where  $\overline{\phi} = 0$ which gives:

$$  \overline{\phi} = \frac{\phi_0 y_0}{y}(1-exp(-y/y_0))$$

***

The average bulk density, likewise, is a modification of the above 2 equations, assuming that the density of the sediment framework does not vary, it is given by

$$  \overline{\rho_b} = \overline{\phi} \rho_w  + (1-\overline{\phi})\rho_s $$ 

Using the following parameter values: 
+ $\rho_w=1000\, kg\,m^{-3}$,  
+ $\rho_s=2650\, kg\,m^{-3}$, 
+ $\phi_0 = 0.6$ and 
+ $y_0 = 1350\,m$.
